```python
import pandas as pd

data = pd.read_csv('predicción_diaria_transform_2.csv')
```

```python
 data
```

```python
didx = pd.DatetimeIndex(data['UTC time']) 
data = data.set_index(didx)
data = data.drop(['UTC time', 'Hora'], axis = 1)
  
```

```python
 data  =  data['2021-10':]
```

```python
data
```

```python
from pycaret.time_series import *
setup(data, fh = 150, fold = 2, session_id = 123, use_gpu = True )
```

```python
check_stats()
```

```python
plot_model(plot = 'cv')
```

```python
plot_model(plot = 'acf')
```

```python
# Diagnostics plot
plot_model(plot = 'diagnostics')
```

```python
# Decomposition plot
plot_model(plot = 'decomp_stl')
```

```python
best = compare_models()
```

```python
pred_holdout = predict_model(best)
```

```python
# create fbprophet model
snaive = create_model('snaive')
print(snaive)
```

```python
tuned_snaive = tune_model(snaive)
print(tuned_snaive)
```

```python
plot_model(best, plot = 'forecast')
```

```python
# forecast in unknown future
plot_model(best, plot = 'forecast', data_kwargs = {'fh' : 360})
```

```python
plot_model(tuned_snaive, plot = 'forecast', data_kwargs = {'fh' : 400})
```
